## Edges & Gradients
- Find horizontal and vertical gradients using Sobel
- Create edge mask using Canny
- Show the magnitude of the gradients

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("../Images/image2.jpg")
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Blurring image with kernel
k = np.array([[1,4,1], 
              [4,7,4], 
              [1,4,1]])/27
blurred = cv2.filter2D(grayImage, ddepth=-1, kernel=k)

# Using Sobel to find horizontal and vertical gradients
Ix = cv2.Sobel(blurred, ddepth=cv2.CV_64F, dx=1, dy=0)
Iy = cv2.Sobel(blurred, ddepth=cv2.CV_64F, dx=0, dy=1)
```

```python
# Creating edge mask using Canny (255 for edges 0 for not)
E = cv2.Canny(img, threshold1 = 255, threshold2 = 0)
```

```python
# On the edges, show the magnitude of the gradients
mag = cv2. magnitude(Ix, Iy) 
```

```python
# View images
plt.figure(figsize=(10, 10))

plt.subplot(2,2,1)
plt.imshow(Ix, cmap="gray")
plt.title('Ix')

plt.subplot(2,2,2)
plt.imshow(Iy, cmap="gray")
plt.title('Iy')

plt.subplot(2,2,3)
plt.imshow(E, cmap="gray")
plt.title('Edge')

plt.subplot(2,2,4)
plt.imshow(mag, cmap="gray")
plt.title('Magnitude')
```
