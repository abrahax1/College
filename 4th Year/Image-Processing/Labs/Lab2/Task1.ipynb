## Access Pixels
- Opening cmages
- Changing colour
- Converting colour
- Modifiying *Luminance* and *Intensity*
- Showing two images in one matplot

```python
# Read in any image
import cv2 
import easygui
from matplotlib import pyplot as plt

I = cv2.imread("../Images/image2.jpg")
```

```python
# Colour change
I[:200,:200] = [255,255,255]
cv2.imshow("image", I)

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert this image to YUV
yuv = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
cv2.imshow("image2", yuv)

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Extract the Y channel (luminance)
no_luminance = yuv

cv2.imshow("image", no_luminance[:,:,0])

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert image to grayscale by changing the intesity 
no_intensity = yuv

cv2.imshow("image", no_intensity[:,:,1:2])

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert image to grayscale by changing the intesity 
no_intensity = yuv

cv2.imshow("image", no_intensity[:,:,1])

key = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Display two images side by side using matplot
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(yuv)
plt.title('YUV')

plt.subplot(2,2,2)
im = cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB)
plt.imshow(im)
plt.title('RGB')
```

```python
# Show the luminance on the screen side by side with the intensity
# using MatPlotLibs subplot function
plt.subplot(2,2,1)
plt.imshow(yuv[:,:,0])
plt.title('Luminance')

plt.subplot(2,2,2)
plt.imshow(yuv[:,:,1:2])
plt.title('Intensity')
```
