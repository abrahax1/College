## Histogram Tasks
- Creating a histogram
- Using equalization to brighten the image

```python
# Open image “Wartime.jpg”
import cv2 
from cv2 import cvtColor 
import easygui
from matplotlib import pyplot as plt

I = cv2.imread("../Images/wartime.jpg")
I = cvtColor(I, cv2.COLOR_BGR2RGB)
```

```python
# View its histogram alongside the image
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(I)
plt.title('Image')

Values = I.ravel()
plt.subplot(2,2,2)
plt.hist(Values,bins=256,range=[0,256])
plt.title('Histogram')
```

```python
# Use histogram equalisation to improve its contrast
plt.figure(figsize=(10,10))
grayImage = I
grayImage = cv2.cvtColor(I, cv2.COLOR_RGB2GRAY)
grayImage = cv2.equalizeHist(grayImage)

Values = I.ravel()
Values1 = grayImage.ravel()

plt.subplot(2,2,1)
plt.imshow(I)
plt.title('Image')

plt.subplot(2,2,2)
plt.hist(Values,bins=256,range=[0,256])
plt.title('Hist1')

plt.subplot(2,2,3)
grayImage = cv2.cvtColor(grayImage, cv2.COLOR_GRAY2BGR)
plt.imshow(grayImage)
plt.title('Equalized')

plt.subplot(2,2,4)
plt.hist(Values1,bins=256,range=[0,256])
plt.title('Hist2')
```
