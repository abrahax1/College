## Image adjusting
- Scaling 
- Cropping
- Rotating

```python
import cv2
from cv2 import cvtColor
from matplotlib import pyplot as plt

I = cv2.imread("../Images/wartime.jpg")
I = cvtColor(I, cv2.COLOR_BGR2RGB)
```

```python
# Crop out the fourth quadrant of the image

# [bottom to top / top to bottom : right to left / left to right]
C = I[300:,150:]
# C[300:,150:] = [0,0,0]
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(I)
plt.title('Image')

plt.subplot(2,2,2)
plt.imshow(C)
plt.title('Image')
```

```python
# Rotate the cropped section by 45
height, width = C.shape[:2]
cx, cy = (height // 2, width // 2)

matrix = cv2.getRotationMatrix2D(center = (cx, cy), angle = 45, scale = 1.0)
R = cv2.warpAffine(C, M = matrix, dsize = (width, height))

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(R)
plt.title('Image')
```

```python
# Scale the image
S = cv2.resize(I, dsize=(3 * width, 4 * height))

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(S)
plt.title('Image')
```
