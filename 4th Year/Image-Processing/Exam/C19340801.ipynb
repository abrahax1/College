### get_dimensions function
- Inside this function I have created multiple lists to store the Left and Right dimensions.
- I also used an adaptative threshold and then masked the image.
- Then I Inverted the mask to find the area of interest.

### bothSides function
- I created this function to find the contours of the inverse_mask.
- Then found the 3 biggest contours.
- I use a loop to go through these and check the height and width:
    - To find the left object I check that the height is less than 309 and width less than 200.
    -  I draw the rectangle for the object matching these values and append the values to the lists created in the first function.
- The process is repeated for the right object but this time checking that:
    - The height is more than 350 and width more than 200.
    - Once again draw the rectangle on the object and append the values to the lists created in the first function.
- This function returns the LeftDimensions and RightDimensions containing:
    - The height and width for both objects 

### nicePrint function 
- This function is to append the correct values to the width and height:
- Also it converts the pixel values to cm, by dividing the value by 26.4 (according to online conversion)

### Values meaning  
- The way it is displaying is: "[2.0, 'cm', 9.0, 'cm'] height: [11.0, 'cm', 21.0, 'cm']"
- This means:
    - Left object: width is 2.0cm and height is 11.0cm 
    - Right object: width is 9.0cm and height is 21.0cm 


```python
import cv2
import numpy as np


def get_dimensions(image_path):
    LeftWidth = []
    LeftHeight = []
    LeftDimension = []

    RightWidth = []
    RightHeight = []
    RightDimension = []
    I = cv2.imread(image_path)
    
    grayImage = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
    Mask = cv2.adaptiveThreshold(grayImage, maxValue = 255,adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,thresholdType = cv2.THRESH_BINARY,blockSize = 205, C = 8)
    inverse_mask = cv2.bitwise_not(Mask)
    ROI = cv2.bitwise_and(I, I, mask = inverse_mask)

    def bothSides():
        contours, hierarchy = cv2.findContours(inverse_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        cnts = sorted(contours, key = cv2.contourArea, reverse = True)[:3]

        for c in (cnts):
            x,y,width,height = cv2.boundingRect(c)
            if height < 309 and width < 200:
                cv2.rectangle(ROI,(x,y),(x+width,y+height),(255,255,0),2)
                x,y,width,height = cv2.boundingRect(c)
                LeftWidth.append(width)
                LeftHeight.append(height)

                LeftDimension.append(LeftWidth)
                LeftDimension.append(LeftHeight)

                cv2.imshow('txt', ROI)
                cv2.waitKey(0)
                cv2.destroyAllWindows()

            if height > 350 and width > 200:
                cv2.rectangle(ROI,(x,y),(x+width,y+height),(0,0,255),2)
                x,y,width,height = cv2.boundingRect(c)
                RightWidth.append(width)
                RightHeight.append(height)

                RightDimension.append(RightWidth)
                RightDimension.append(RightHeight)


        return(LeftDimension, RightDimension)


    def nicePrint(LeftDimension, RightDimension):
        width = []
        width.append(LeftDimension[0])
        width.append(RightDimension[0])
        newWidth = []
        newHeight = []
        for i in range(len(width)):
            newWidth.append(width[i][0] // 26.4)
            newWidth.append('cm')

        height = []
        height.append(LeftDimension[1])
        height.append(RightDimension[1])

        for i in range(len(height)):
            newHeight.append(height[i][0] // 26.4)
            newHeight.append('cm')

        return newWidth,newHeight

    LeftDimension, RightDimension = bothSides()
    width, height = nicePrint(LeftDimension, RightDimension)
    # width = 0
    # height = 0
    
    # calculate width and height and assign them to the width and height variables
    return width, height


box_width, box_height = get_dimensions('SBox.jpg')
print(f"width: {box_width} height: {box_height}")

gun_width, gun_height = get_dimensions('SGun.jpg')
print(f"width: {gun_width} height: {gun_height}")
```
